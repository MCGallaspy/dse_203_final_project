```python
import json
import pprint
import re
import statistics
import RAKE
import operator


# Raksetup with stopword directory
stop_dir = "SmartStoplist.txt"
rake_object = RAKE.Rake(stop_dir)

with open('updated_scraped_data.json') as f:
  data = json.load(f)

#pprint.pprint(len(data.keys()))

interest = ['SlideShare','Glassdoor','Lynda Weinman']

for companykey in data.keys():
    if companykey in interest:
        IBM_json = data[companykey]
        for key in IBM_json.keys():
            if '[edit]' in key or 'Intro' in key:
                if 'Intro' in key:
                    IBM_json[key] = [IBM_json[key]]
                print ("**********"+key)
                for item in IBM_json[key]:
                    if isinstance(item,str):
                        item_ = item.replace("\n","")
                        regex = '\[.*?\]'
                        item_ = re.sub(regex,'',item_)
                        keywords = rake_object.run(item_)
                        print ("keywords: ", keywords)
                IBM_json[key].append(keywords)                
        with open(companykey, "w+") as f:
            f.write(json.dumps(IBM_json, indent=4))
```
