```python
import json
import pprint

with open('upd_json__.json') as f:
    data = json.load(f)
```

```python
verb_set = []
for companykey in data.keys():
    print (companykey)
    if "IDA Ireland" in companykey:
        continue
    else:
        GD_json = data[companykey]
        for key in GD_json:
            if ('edit' in key or 'Intro' in key) and ('References' not in key) and ('See also' not in key) and ('External links' not in key):
                print ('      '+key)
                if isinstance (GD_json[key][2],dict):
                    pos = GD_json[key][2]
                elif isinstance (GD_json[key][3],dict):
                    pos = GD_json[key][3]
                else:
                    continue
                for sentence in pos.keys():
                    for item in pos[sentence].replace("(","").split(')'):
                        if len(item.split(','))>1:
                            if 'VB' in item.split(',')[1]:
                                #print (item.split(',')[0])
                                verb_set.append(item.split(',')[0])
print(len(verb_set))
```

```python
from nltk.stem import LancasterStemmer
lancaster=LancasterStemmer()

stem_set = []
for w in verb_set:
    stem_set.append(lancaster.stem(lancaster.stem(w.replace("'",""))))
    
hash_map = {}
for word in stem_set:        
    if word in hash_map:
        hash_map[word] = hash_map[word] + 1
    else:
        hash_map[word] = 1
import operator
sorted_hashmap = sorted(hash_map.items(), key=operator.itemgetter(1),reverse=True)  
pprint.pprint(sorted_hashmap)
```
