```python
import json
import pprint
import pandas as pd

with open('updated_scraped_data.json') as f:
    data = json.load(f)

skip_f = open('skiplist2.txt', "r")
skip_list = []
for line in skip_f:
    skip_list.append(line.strip())
```

```python
#Extract all tables from the Input Big json file
df_list = []
for companykey in data.keys():
    if (companykey not in skip_list) and ("Authority File" not in companykey):
        #print (companykey)
        GD_json = data[companykey]        
        for key in GD_json:
            flag =0
            if ('[edit]' in key) and ('External links' not in key):
                #print ('\t' + str(key))
                for item in GD_json[key]:
                    if isinstance (item,list):
                        for item_ in item:
                            if (isinstance (item_,dict)) and (item_):
                                flag = 1
                        if flag==1:
                            if (input('Wanna see DF of ' + companykey + '\n\t' + str(key) + '?\t')=='y'):
                                df = pd.DataFrame(item)
                                display (df)
                                if (input('save this?\t')=='y'):
                                    print ('Alrite')
                                    df_list.append(companykey)
                                    df_list.append(key)
                                    df_list.append(df)
                                    print (len(df_list))
                                    flag =0
                            else:
                                flag =0
```

```python
import re
import csv

def create_oneway_edges(company,section,info_df,edge_id):
    c = open("connectivity.tsv", "a")
    l = open("edgelabel.tsv", "a")
    p = open("edgeproperty.tsv", "a")
    conn  = csv.writer(c,delimiter='\t')
    label = csv.writer(l,delimiter='\t')
    props = csv.writer(p,delimiter='\t')
    info_list = list(info_df.columns)
    info_list.append(company)
    info_list.append(section)
    print (info_list)
    for index,item in enumerate(info_list):
        print (index,item)
    fn,el,en = input("Hmm..").split(',')
    print (fn)
    print (en)
    if info_list[int(fn)]:
        if info_list[int(fn)] in info_df.columns:
            fn_4m_df = 1
            fromnode = info_list[int(fn)]
        else:
            fn_4m_df = 0
            fromnode = info_list[int(fn)]
    else:
        fromnode = fn
        fn_4m_df = 0
    if info_list[int(en)]:
        if info_list[int(en)] in info_df.columns:
            en_4m_df = 1
            endnode = info_list[int(en)]
        else:
            en_4m_df = 0
            endnode = info_list[int(en)]
    else:
        endnode = en
        en_4m_df = 0
    props_list=[]
    props_name_list = []
    rsp = input("Properties?\t") 
    while (rsp!='n'):
        props_list.append(info_list[int(rsp)])
        props_name_list.append(input('Prop Name?\t'))
        rsp = input("More Properties?\t")
    for row in info_df.iterrows():
        edge_id = edge_id+1;
        if fn_4m_df==1:
            ifn = row[1][fromnode]
        else:
            ifn = fromnode
        if en_4m_df==1:
            ien = row[1][endnode]
        else:
            ien = endnode
        print ([str(edge_id),ifn, ien])
        print ([str(edge_id),el])            
        conn.writerow([str(edge_id),ifn, ien])
        label.writerow([str(edge_id),el])
        if props_list:
            for idx,item in enumerate(props_list):
                regex = '\[.*?\]'
                val = row[1][item].replace("-","")
                val = re.sub(regex,'',val)
                if val!= "":
                    props.writerow([str(edge_id),props_name_list[idx],row[1][item]])
                    #print ([str(edge_id),props_name_list[idx],row[1][item]])
    c.close()
    l.close()                       
    p.close()
    return edge_id
```

```python
node_prop = {};
edge_id = 1419;
for i in range(int(len(df_list)/3)):
    company = df_list[(i*3)]
    section = df_list[(i*3)+1]
    info_df = df_list[(i*3)+2]
    display (company)
    display (section)
    display (info_df)
    rsp = input('Wanna create edges?\t')
    if (rsp =='y'):
        edge_id = create_oneway_edges(company,section,info_df,edge_id)
    elif(rsp == 'n'):
        mydict = {}
        mydict[section] = info_df.to_dict('dict')
        node_prop[company] = mydict
    else:
        pass
```
